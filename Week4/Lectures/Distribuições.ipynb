```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as sct
```

```python
norm = sct.norm.rvs(loc=10, scale=3, size=1000)

sns.distplot(norm)
```

```python
norm.mean(), norm.std()
```

```python
sct.norm.cdf(8, loc=10, scale=3)
```

```python
sct.norm.cdf(10, loc=5, scale=3)
```

```python
x = np.linspace(0, 20, 1000)

cdf = sct.norm.cdf(x, 10, 3)


sns.lineplot(x, cdf)
```

```python
sct.norm.sf(8, 10, 3) #surviver function
```

```python
sct.norm.ppf(0.1, loc=10, scale=3) # Probabilidade acumulada, entrada com dado em %
```

```python
sct.norm.isf(0.25, 10, 3) # Lado direito
```

```python
sct.norm.pdf(10, loc=10, scale=3) # Densisdade de probabi
```

# Distribuições binomiais

```python
binom_20_05 = sct.binom.rvs(20, 0.5, size=10000) # Média 10, variança 5
binom_20_07 = sct.binom.rvs(20, 0.7, size=10000) # Média 14, Variância 4.2

sns.distplot(binom_20_05,bins=range(4, 18), kde=False)
sns.distplot(binom_20_07, bins=range(8, 22), kde=False)
```
